**Explanatory Data Analysis, for Chicago Crime Data**

Put the path for the specific dataset under pd.read_csv('')
on skipping bad values use on_bad_lines='skip'

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
df1 = pd.read_csv('/content/drive/MyDrive/Datasets/crimedata/Chicago_Crimes_2001_to_2004.csv', on_bad_lines='skip')
df2 = pd.read_csv('/content/drive/MyDrive/Datasets/crimedata/Chicago_Crimes_2005_to_2007.csv', on_bad_lines='skip')
df3 = pd.read_csv('/content/drive/MyDrive/Datasets/crimedata/Chicago_Crimes_2012_to_2017.csv', on_bad_lines='skip')
df4 = pd.read_csv('/content/drive/MyDrive/Datasets/crimedata/Chicago_Crimes_2008_to_2011.csv', on_bad_lines='skip')
```

Concatenation of the csv files to create 1 file for reading and calling the dataframe as df again to show the values stored in the csv.

```python
df = pd.concat([df1, df2, df3, df4])
dfa=df.head()
dfa
```

Checks all the missing values present in the concatenated dataset

```python
missing_values = df.isnull().sum()
total_values = len(df)
missing_percentages = (missing_values / total_values) * 100

for column, percentage in zip(df1.columns, missing_percentages):
    print(f"{column}: {percentage:.3f}% missing values")
```

Removes null values which is a very important task while preprocessing a dataset

```python
#cleaning dataset
df= df.dropna()
```

Takes the sum of missing values in per category and calculates the percentage of missing values/total values which is supposed to be null the second time


```python
missing_values = df.isnull().sum()
total_values = len(df)
missing_percentages = (missing_values / total_values) * 100

for column, percentage in zip(df1.columns, missing_percentages):
    print(f"{column}: {percentage:.3f}% missing values")
```
